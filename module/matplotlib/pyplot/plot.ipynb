## 1. 默认x为range(len(y))

```python
%matplotlib inline
from matplotlib import pyplot as plt
x =range(10)
y = [2*i +1 for i in x]
plt.plot(y)
```

```python
plt.plot(x,y)
```

## 2. 绘画多个图时直接多次写，此时不能省略x
plot(x1,y1,x2,y2)

```python
plt.plot(x,y,x,x)
```

## 3. 线还可以加上颜色和类型 

```python
colors = ['b','g','r','c','m','y','k','w']
for i,c in enumerate(colors):
    plt.subplot(int('42'+str(i+1)))
    plt.plot(x,x,c)
                
```

In addition, you can specify colors in many weird and wonderful ways, including full names ('green'), hex strings ('#008000'), RGB or RGBA tuples ((0,1,0,1)) or grayscale intensities as a string ('0.8'). Of these, the string specifications can be used in place of a fmt group, but the tuple forms can be used only as kwargs.

```python
colors = ['blue','#001234',(0.5,0.5,0.4),'0.7']
for i,c in enumerate(colors):
    plt.subplot(int('22' + str(i+1)))
    plt.plot(x,x,color=c)
```

线的类型也可以指定

```python
typeLine = ['-','--','-.',':']
print(len(typeLine))
for i,typeline in enumerate(typeLine):
    plt.subplot(2,2,i+1)
    plt.plot(x,x,typeline)
```

```python
typeLine = ['.',',','o','v','^','<','>','1','2','3','4','s','p','*','h','H','+','x','D','d','|','_']
print(len(typeLine))
for i,typeline in enumerate(typeLine):
    plt.subplot(4,7,i+1)
    plt.plot(x,x,typeline)
```

```python
plt.plot(x,x,lineWidth=10)
```
