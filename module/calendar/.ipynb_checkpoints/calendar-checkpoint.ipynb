# 1. calendar.isleap(year)
判断year是否为闰年。

```python
print(calendar.isleap(0))
print(calendar.isleap(2000))
print(calendar.isleap(2100))
print(calendar.isleap(-4))
```

# 2. calendar.leapdays(y1,y2)
判断[y1,y2)之间的闰年数

```python
print(calendar.leapdays(2004,2004))
print(calendar.leapdays(2000,2222))
print(calendar.leapdays(2000,3000))
```

# 3. calendar.monthrange(year,momth)
返回(year,month)的二元组，该月份第一天星期几(0~6)，及这个月的天数。

```python
print(calendar.monthrange(2017,9))
```

# 4. calendar.weekday(year, month, day)

```python
calendar.weekday(2017,8,4)
```
